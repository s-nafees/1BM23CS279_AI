```python
import math

# Initial values for alpha and beta
# Alpha: The best (highest-value) choice found so far for the Maximizing player.
# Beta: The best (lowest-value) choice found so far for the Minimizing player.
MAX_INF = math.inf
MIN_INF = -math.inf

def minimax(depth, nodeIndex, maximizing_player, values, alpha, beta):
    """
    Implements the Alpha-Beta Pruning algorithm.
    
    Args:
        depth (int): Current depth in the game tree.
        nodeIndex (int): Index of the current node's value in the 'values' array.
                         (Assumes a binary tree structure for simplicity).
        maximizing_player (bool): True if it's the Maximizing player's turn, False for Minimizing player.
        values (list): List of heuristic values for the leaf nodes.
        alpha (float): The best value found so far for the Maximizer on the path to the root.
        beta (float): The best value found so far for the Minimizer on the path to the root.
        
    Returns:
        float: The optimal value for the current player at the current node.
    """
    
    # ðŸ›‘ Base Case: Terminating condition (e.g., maximum depth reached or game over)
    # The 'values' array is assumed to hold the leaf node scores.
    # Assumes a full binary tree of depth 3, so a leaf is reached when depth is 3.
    MAX_DEPTH = 3 
    if depth == MAX_DEPTH:
        # nodeIndex is used to access the leaf node score
        return values[nodeIndex]

    # --- Maximizing Player's Turn ---
    if maximizing_player:
        best_val = MIN_INF
        
        # Iterate over children (assuming two children per internal node: index*2 and index*2 + 1)
        for i in range(2):
            child_index = nodeIndex * 2 + i
            
            # Recursive call
            val = minimax(depth + 1, child_index, False, values, alpha, beta)
            best_val = max(best_val, val)
            alpha = max(alpha, best_val)
            
            # âœ‚ï¸ Alpha-Beta Pruning Condition (Beta Cutoff)
            # If current best_val for MAX is >= than the best MIN has found so far (beta),
            # then MIN will never choose this path, so we can stop exploring this branch.
            if beta <= alpha:
                break
                
        return best_val

    # --- Minimizing Player's Turn ---
    else:
        best_val = MAX_INF
        
        # Iterate over children
        for i in range(2):
            child_index = nodeIndex * 2 + i
            
            # Recursive call
            val = minimax(depth + 1, child_index, True, values, alpha, beta)
            best_val = min(best_val, val)
            beta = min(beta, best_val)
            
            # âœ‚ï¸ Alpha-Beta Pruning Condition (Alpha Cutoff)
            # If current best_val for MIN is <= than the best MAX has found so far (alpha),
            # then MAX will never choose this path, so we can stop exploring this branch.
            if beta <= alpha:
                break
                
        return best_val

# --- Example Usage ---
if __name__ == '__main__':
    # Leaf node values (must match the expected number of leaf nodes for the assumed tree structure)
    # Example tree: Root (Depth 0) -> 2 Children (Depth 1) -> 4 Grandchildren (Depth 2) -> 8 Leaves (Depth 3)
    values = [10, 9, 14, 18, 5, 4, 50, 3] 
    
    # Initial call: Start at depth 0, nodeIndex 0, maximizing player's turn, with initial alpha and beta
    optimal_value = minimax(0, 0, True, values, MIN_INF, MAX_INF)
    
    print(f"The optimal value for the Maximizing Player is: {optimal_value}")
```
